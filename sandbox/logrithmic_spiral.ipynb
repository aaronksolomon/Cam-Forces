```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import init_printing
from IPython.display import Math, display

np.set_printoptions(precision=2)

init_printing()

# Parameters
r_0 = 1  # Base radius
alpha = np.radians(15)  # Angle in degrees, converted to radians
phi = np.linspace(0, np.pi/2, 1000)  # Range of 

# Logarithmic spiral function
r_vals = r_0 * np.exp(np.tan(alpha) * phi)

# Plotting in polar coordinates
plt.figure(figsize=(6, 6))
ax = plt.subplot(111, projection='polar')
ax.plot(phi, r_vals)

# Customize the plot (optional)
ax.set_title(r"Logarithmic Spiral $r(\phi) = r_0 \cdot e^{\tan(\alpha) \cdot \phi}$")
ax.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
r_0 = 1  # Base radius
alpha = np.radians(14.5)  # Angle in degrees, converted to radians
offset = 0.14 * r_0  # Offset for the axle point
angle_offset1 = 0.2
angle_offset2 = 0.15
phi = np.linspace(np.pi/2 - angle_offset1, np.pi + angle_offset2, 1000)  # Range of  for 90 degrees

# Logarithmic spiral function
r_vals = r_0 * np.exp(np.tan(alpha) * (phi - np.pi/2)) #start plotting at pi/2 radians, so must offset.

# Convert to Cartesian coordinates for easier plotting
x_spiral = r_vals * np.cos(phi)
y_spiral = r_vals * np.sin(phi)

# Plotting the cam lobe with the rest of the geometry
plt.figure(figsize=(8, 8))

# Shade the area of the cam with grey crosshatching
plt.fill(np.append(x_spiral, offset), np.append(y_spiral, -offset), color='grey', alpha=0.5, hatch='/')

# Plot the logarithmic spiral (cam lobe curve)
plt.plot(x_spiral, y_spiral, label='Cam Lobe Curve')

# Plot the axle circle (centered at origin with radius axle_radius)
axle_radius = r_vals[-1] / 17
axle_circle = plt.Circle((0, 0), axle_radius, color='black', fill=True, linewidth=2)
plt.gca().add_artist(axle_circle)

# Plot the bottom side and vertical right side of the cam lobe
plt.plot([x_spiral[-1], offset], [y_spiral[-1], -offset], 'k-', lw=2)

# Plot the right side of the cam lobe (vertical line at x = offset)
plt.plot([offset, x_spiral[0]], [-offset, y_spiral[0]], 'k-', lw=2, label='Right Edge')

# Customize the plot
plt.title('Cam Lobe Geometry')
plt.xlabel('X')
plt.ylabel('Y')
# Remove gridlines
#plt.grid(True)
plt.axis('equal')
plt.legend()

plt.show()
```

```python
import sympy as sp

# Define symbolic variables
phi = sp.symbols('phi')
r, r_0, alpha = sp.symbols('r r_0 alpha', real=True, positive=True)

# Define the logarithmic spiral equation
r_exp = r_0 * sp.exp(sp.tan(alpha) * (phi - sp.pi/2))

r_eq = sp.Eq(r, r_exp)
# Display the equation
display(r_eq)

# Compute the derivative of r with respect to phi
dr_dphi = sp.diff(r_exp, phi)
display(Math(r"\text{Derivative } \frac{d}{d\phi}"))
display(dr_dphi)

# Compute the Cartesian coordinates
x_spiral = r_exp * sp.cos(phi)
y_spiral = r_exp * sp.sin(phi)

# Display the Cartesian coordinates
print("cartesian coordinate expressions (x, y):")
display(x_spiral, y_spiral)
```
